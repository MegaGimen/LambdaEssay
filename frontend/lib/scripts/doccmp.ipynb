```python
from docx2python import docx2python
import difflib

def flatten_body(body):
    lines = []
    for section in body:
        for table in section:
            for row in table:
                for cell in row:
                    text = ''.join(str(s) for s in cell if s.strip() or 'media/' in str(s))
                    lines.append(text)
    return lines

# 加载文档
doc1 = docx2python("../1.docx")
doc2 = docx2python("../2.docx")

# 扁平化
lines1 = flatten_body(doc1.body)
lines2 = flatten_body(doc2.body)

# 生成 diff
diff = difflib.unified_diff(lines1, lines2, fromfile='1.docx', tofile='2.docx', lineterm='')
print('\n'.join(diff))
```
